```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

<h1> For which superpopulation our model makes the best and the worse predictions?</h1>
For that we did a single random sample test on a 1000 models created from random samples and placed them in bins for their
average errors on each of the superpopulations.

```python
data = pd.read_csv('./generated/per_pop_error(2).csv')

np_data = np.array([
    data["AFR Mean Error Rate On Random Samples"].to_numpy(),
    data["EUR Mean Error Rate On Random Samples"].to_numpy(),
    data["EAS Mean Error Rate On Random Samples"].to_numpy(),
    data["AMR Mean Error Rate On Random Samples"].to_numpy(),
    data["SAS Mean Error Rate On Random Samples"].to_numpy()
])

np_data = np.rot90(np_data)

data.head()
```

```python
bins = [10, 20, 50, 100, 200]

for bin in bins:
    colors =  ['#D9EDF8', '#FFADAD', '#E4F1EE', '#FFD6A5', '#DEDAF4']
    plt.hist(np_data, bin, histtype='bar', color=colors, stacked=True, fill=True, label=["AFR", "EUR", "EAS", "AMR", "SAS"])
    plt.legend()
    plt.xlabel(xlabel="Error Rate")
    plt.ylabel(ylabel="Number of Simulations")
    plt.show()
```
